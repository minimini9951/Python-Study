## Bar Chart
- bar
- barh
- http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.bar
- http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.barh

```python
y = [2, 3, 1]
x = np.arange(len(y))

plt.title('Bar Chart')
xlabel = ['가', '나', '다']
plt.bar(x, y)
plt.xticks(x, xlabel)  # x 눈금 값 변경
plt.yticks(sorted(y))  # y 눈금 값 변경
plt.xlabel('가나다')  # x 축 제목변경
plt.ylabel('빈도수')  # y 축 제목변경
plt.show()
```

```python
np.random.seed(0)

people = ['몽룡', '춘향', '방자', '향단']
y_pos = np.arange(len(people))
data = np.random.rand(len(people)) * 10
error = np.random.rand(len(people))

plt.title('수평 막대 그래프')
plt.barh(y_pos, data, xerr=error, alpha=0.4)
plt.yticks(y_pos, people)
plt.xlabel('x 라벨')
plt.show()
```

```python
# stem plot : bar chart와 비슷하지만 폭이 없는 그래프
# 주로 이산확률함수나 자기상관관계를 묘사할 때 사용
# http://matplotlib.org/api

x = np.linspace(0.1, 2 * np.pi, 10)
plt.title('Stem Plot')
plt.stem(x, np.cos(x), '-.')
plt.show()
```

## pie chart
- http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.pie
- 카테고리별 값의 상대적인 비교를 해야할 때 주로 사용
- 파이차트를 그릴 때 원의 형태를 유지할 수 있도록 plt.asix('equal') 명령어를 실행

```python
labels = ['맥주', '소주', '양주', '막걸리']
sizes = [25, 45, 10, 20]
colors = ['yellowgreen', 'lightskyblue', 'gold', 'lightcoral']
explode = (0, 0.1, 0, 0)

plt.title('내가 간절히 원하는 것은?')
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%',
        shadow=True, startangle=0)
plt.axis('equal')
plt.show()
```

## scatter plot : 산점도(산포도)
- 2차원 데이터인 연속형 데이터 집합의 상관관계 파악용 도구
- 점 하나의 위치(x, y)의 값 필요

```python
np.random.seed(0)

x = np.random.normal(0, 1, 100)
y = np.random.normal(0, 1, 100)

plt.title('Scatter Plot')
plt.scatter(x, y)
plt.show()
```

```python
# bubble chart : 2차원이 아니라 3차원 또는 4차원의 산점도를 표현할 경우 점하나의 크기 혹은 색을 이용하여 다른 데이터를 표시함
# s : 크기, c : 색깔
N = 30
np.random.seed(0)

x = np.random.rand(N)
y1 = np.random.rand(N)
y2 = np.random.rand(N)
y3 = np.pi * (15*np.random.rand(N)) ** 2

plt.title('Bubble Chart')
plt.scatter(x, y1, c=y2, s=y3)
plt.show()
```

## Imshow() : 화상(image)데이터를 표현하는 시각화 도구
- 이미지데이터도 행과 열을 가진 2차원 데이터이다.

```python
from sklearn.datasets import load_digits
digits = load_digits()
digits
```

```python
x = digits.images[0]
x
```

```python
plt.title('손글씨 데이터셋(mnist digits : 0)')
plt.imshow(x, interpolation='nearest', cmap=plt.cm.bone_r)

plt.xticks()
plt.yticks()
plt.grid(False)
plt.subplots_adjust(left=0.35, right=0.65, bottom=0.35, top=0.65)
plt.show()
```

```python
methods = ['none', 'nearest', 'bilinear', 'bicubic', 'spline16', 'spline36',
           'hanning', 'hamming', 'hermite', 'kaiser', 'quadric', 'catrom',
           'gaussian', 'bessel', 'mitchell', 'sinc', 'lanczos']

fig, axes = plt.subplots(3, 6, figsize=(12, 6),
                         subplot_kw={'xticks': [], 'yticks': []})

for ax, interp_method in zip(axes.flat, methods):
    ax.imshow(x, interpolation=interp_method, cmap=plt.cm.bone_r)
    ax.set_title(interp_method)

plt.show()
```
