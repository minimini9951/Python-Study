### mathplotlib 라이브러리로 그래프 그리는 순서
- 1. 전체 그래프가 위치할 기본 틀 만들기
- 2. 그래프를 그려 넣을 그래프 격자 만들기
- 3. 격자에 그래프를 하나씩 추가 : 그래프는 왼쪽에서 오른쪽으로 추가됨
- 4. 격자가 꽉차면 다음 행에 그래프를 그려넣음

#### 한 번에 4개의 그래프 그리기

```python
# 데이터를 불린형으로 추출하여 데이터셋에 저장
dataset_2 = anscombe[anscombe['dataset'] == 'II']
dataset_3 = anscombe[anscombe['dataset'] == 'III']
dataset_4 = anscombe[anscombe['dataset'] == 'IV']
```

```python
fig = plt.figure()
```

##### 그래프 격자 그리기 : add_subplot()

```python
axis1.plot(dataset_1['x'], dataset_1['y'], 'o')
axis2.plot(dataset_2['x'], dataset_2['y'], 'o')
axis3.plot(dataset_3['x'], dataset_3['y'], 'o')
axis4.plot(dataset_4['x'], dataset_4['y'], 'o')

fig
```

```python
axis1.set_title('dataset_1')
axis2.set_title('dataset_2')
axis3.set_title('dataset_3')
axis4.set_title('dataset_4')

fig
```

## 기초 그래프 그리기 : 히스토그램, 산점도, 박스
- 히스토그램 : 데이터 분포와 빈도를 살펴보는 용도

### tips 데이터셋 불러와서 저장

```python
tips = sns.load_dataset('tips')
print(tips.head())
print(type(tips))
```

### 히스토그램 그리기

#### 기본 틀 생성

```python
fig = plt.figure()
axes1 = fig.add_subplot(1, 1, 1)
```

#### 히스토그램 그리기 : hist()

```python
axes1.hist(tips['total_bill'], bins=10)
axes1.set_title('Histogram of Total Bill')
axes1.set_xlabel('Frequency')
axes1.set_ylabel('Total Bill')

fig
```

### 산점도 그래프 : scatter()

```python
scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1, 1, 1)
axes1.scatter(tips['total_bill'], tips['tip'])
axes1.set_title('Scatterplot of Total Bill vs Tip')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')
```

### 박스 그래프 : boxplot()

```python
boxplot = plt.figure()
axes1 = boxplot.add_subplot(1, 1, 1)

axes1.boxplot([tips[tips['sex'] == 'Female']['tip'],
              tips[tips['sex'] == 'Male']['tip']],
              labels=['Female', 'Male'])

axes1.set_xlabel('Sex')
axes1.set_ylabel('Tip')
axes1.set_title('Boxplot of Tips by Sex')
```

## 다변량 그래프
- 3개이상의 변수를 사용한 그래프

### 산점도 그래프 : 성별 변수추가(색상으로 구분)

```python
# 성별을 1과 0으로 구분하는 함수 생성
def record_s(s):
    if s == 'Female':
        return 0
    else:
        return 1
```

```python
# 색상으로 구분하기위해 0과 1로 구분된 열 추가
tips['sex_color'] = tips['sex'].apply(record_s)
```

```python
scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1, 1, 1)
axes1.scatter(
    x=tips['total_bill'],
    y=tips['tip'],
    s=tips['size'] * 10,
    c=tips['sex_color'],
    alpha=0.5)
axes1.set_title('Total Bill vs Tip Colored by Sex and Sized by Size')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')
```

## 단변량 그래프

### 히스토그램

```python
import seaborn as sns
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
ax = plt.subplots()
ax = sns.histplot(tips['total_bill'], kde=True)
ax.set_title('Total Bill Histogram')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Frequency')
```

```python
tips.head()
```

```python
ax = plt.subplots()
ax = sns.countplot('day', data=tips)
ax.set_title('Count of Days')
ax.set_xlabel('Day of the Week')
ax.set_ylabel('Frequency')
```

### 산점도 그래프 : scatter()

#### 회귀선 함께 출력 : regplot()

```python
ax = plt.subplots()
ax = sns.regplot(x='total_bill', y='tip', data=tips)
ax.set_title('Scatterplot of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
```

```python
ax = plt.subplots()
ax = sns.regplot(x='total_bill', y='tip', data=tips, fit_reg=False)
ax.set_title('Scatterplot of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
```

#### 히스토그램 함께 출력 : jointplot()

```python
joint = sns.jointplot(x='total_bill', y='tip', data=tips)
joint.set_axis_labels(xlabel='Total Bill', ylabel='Tip')
joint.fig.suptitle('Joint Plot of Total Bill and Tip', fontsize=10, y=1.03)
```

#### 육각그래프 : hexbim(jointplot과 형식 동일, kind='hex' 지정해주면 변경됨)

```python
hexbin = sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex')
hexbin.set_axis_labels(xlabel='Total Bill', ylabel='Tip')
hexbin.fig.suptitle('Hexbin Joint Plot of Total Bill and Tip',
                    fontsize=10, y=1.03)
```

#### 이차원 밀집도 그래프 : kdeplot()

```python
ax = plt.subplots()
ax = sns.kdeplot(x=tips['total_bill'],
                y=tips['tip'],
                shade=True)
ax.set_title('Kernel Density Plot of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
```

```python
ax = plt.subplots()
ax = sns.barplot(x='time', y='total_bill', data=tips)
ax.set_title('Bar plot of average total bill for time of day')
ax.set_xlabel('Time of day')
ax.set_ylabel('Average total bill')
```

### 박스 그래프 : boxplot()

```python
ax = plt.subplots() # 격자 생성
ax = sns.boxplot(x='time', y='total_bill', data=tips)
ax.set_title('Boxplot of total bill by time of day')
ax.set_xlabel('Time of day')
ax.set_ylabel('Total Bill')
```

#### 바이올린 그래프 : violinplot()
- 박스 그래프에서 데이터 분산이 모호하게 표현될 때 사용

```python
ax = plt.subplot()
ax = sns.violinplot(x='time', y='total_bill', data=tips)
ax.set_title('Violin plot of total bill by time of day')
ax.set_xlabel('Time of day')
ax.set_ylabel('Total Bill')
```

### 관계 그래프 : pairplot
- 중복된 그래프가 출력되기 때문에 원하는 그래프로 교체 해줘야함

```python
fig = sns.pairplot(tips)
```

```python
pair_grid = sns.PairGrid(tips)
pair_gird = pair_grid.map_upper(sns.regplot)
pair_grid = pair_grid.map_lower(sns.kdeplot, shade=True)
pair_grid = pair_grid.map_diag(sns.histplot, kde=True)
plt.show()
```
