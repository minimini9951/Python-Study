- 4. 격자가 꽉차면 다음 행에 그래프를 그려넣음

### tips 데이터셋 불러와서 저장

```python
tips = sns.load_dataset('tips')
print(tips.head())
print(type(tips))
```

### 히스토그램 그리기

#### 기본 틀 생성

```python
fig = plt.figure()
axes1 = fig.add_subplot(1, 1, 1)
```

#### 히스토그램 그리기 : hist()

```python
axes1.hist(tips['total_bill'], bins=10)
axes1.set_title('Histogram of Total Bill')
axes1.set_xlabel('Frequency')
axes1.set_ylabel('Total Bill')

fig
```

### 산점도 그래프 : scatter()

```python
scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1, 1, 1)
axes1.scatter(tips['total_bill'], tips['tip'])
axes1.set_title('Scatterplot of Total Bill vs Tip')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')
```

### 박스 그래프 : boxplot()

```python
boxplot = plt.figure()
axes1 = boxplot.add_subplot(1, 1, 1)

axes1.boxplot([tips[tips['sex'] == 'Female']['tip'],
              tips[tips['sex'] == 'Male']['tip']],
              labels=['Female', 'Male'])

axes1.set_xlabel('Sex')
axes1.set_ylabel('Tip')
axes1.set_title('Boxplot of Tips by Sex')
```

## 다변량 그래프
- 3개이상의 변수를 사용한 그래프

### 산점도 그래프 : 성별 변수추가(색상으로 구분)

```python
# 성별을 1과 0으로 구분하는 함수 생성
def record_s(s):
    if s == 'Female':
        return 0
    else:
        return 1
```

```python
# 색상으로 구분하기위해 0과 1로 구분된 열 추가
tips['sex_color'] = tips['sex'].apply(record_s)
```

```python
scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1, 1, 1)
axes1.scatter(
    x=tips['total_bill'],
    y=tips['tip'],
    s=tips['size'] * 10,
    c=tips['sex_color'],
    alpha=0.5)
axes1.set_title('Total Bill vs Tip Colored by Sex and Sized by Size')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')
```

## Seaborn 라이브러리 사용
- subplots() : figure 와 axes 값을 받음.
- figure : 전체 subplot(서브플로안에 몇개의 그래프가 있던지 상관없이  그걸 담는 하나)
- axes : 전체 내부의 갯수(서브플롯 안에 2개(a1,a2)의 그래프가 있다면 a1, a2)

### 바이올린 그래프 : violinplot() - 색상추가 : (hue = 'x')
- hue : 색상을 다르게 하고싶은 변수이름을 지정

```python
ax = plt.subplots()
ax = sns.violinplot(x='time', y='total_bill', hue='sex', data=tips, split=True)
```

### 산점도, 관계 그래프 그리기
- 색상 추가 : 메서드에 hue 인잣값에 색상에 사용할 열 이름 추가

```python
scatter = sns.lmplot(x='total_bill', y='tip', data=tips,
                     hue='sex', fit_reg=False)
fig = sns.pairplot(tips, hue='sex')
```

#### 크기와 모양, 기호 조절 - 안됨, 선생님한테 물어볼 것
- scatter_kws 에 딕셔너리 형태로 인잣값을 전달

```python
scatter = sns.lmplot(x='total_bill', y='tip', data=tips, fit_reg=False,
                     hue='sex', markers=['o', 'x'],
                     scatter_kws={'s': tips['size']*10})
```

### 4개의 데이터 그룹 그래프 한번에 나타내기 : lmplot()
- col, col_warp 인자를 설정
- fig_reg 인자는 회귀선의 유무를 나타냄

```python
anscombe_plot = sns.lmplot(x='x', y='y', data=anscombe, fit_reg=False,
                           col='dataset', col_wrap=2)
```

#### 번외1. FacetGrid 클래스로 그래프 그리기
- FacetGrid에 데이터프레임과 열이름을 전달하여 객체 생성
- 객체의 map 메서드에 그래프의 종류와 x축으로 지정할 열이름을 순서대로 전달

```python
# FacetGrid에 데이터프레임과 열이름을 전달하여 객체 생성
facet = sns.FacetGrid(tips, col='time')
# 객체의 map 메서드에 그래프의 종류와 x축으로 지정할 열이름을 순서대로 전달
facet.map(sns.histplot, 'total_bill', kde=True)
```

#### 번외2. 다변량 그래프 그리기

```python
facet = sns.FacetGrid(tips, col='day', hue='sex')
facet = facet.map(plt.scatter, 'total_bill', 'tip')
facet = facet.add_legend()
```

#### 번외3. 더 많은 변수사용

```python
facet = sns.FacetGrid(tips, col='time', row='smoker', hue='sex')
facet.map(plt.scatter, 'total_bill', 'tip')
```

### 산점도 그래프 : scatter()

#### 회귀선 함께 출력 : regplot()

```python
ax = plt.subplots()
ax = sns.regplot(x='total_bill', y='tip', data=tips)
ax.set_title('Scatterplot of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
```

```python
ax = plt.subplots()
ax = sns.regplot(x='total_bill', y='tip', data=tips, fit_reg=False)
ax.set_title('Scatterplot of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
```

#### 히스토그램 함께 출력 : jointplot()

```python
joint = sns.jointplot(x='total_bill', y='tip', data=tips)
joint.set_axis_labels(xlabel='Total Bill', ylabel='Tip')
joint.fig.suptitle('Joint Plot of Total Bill and Tip', fontsize=10, y=1.03)
```

#### 육각그래프 : hexbim(jointplot과 형식 동일, kind='hex' 지정해주면 변경됨)

```python
hexbin = sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex')
hexbin.set_axis_labels(xlabel='Total Bill', ylabel='Tip')
hexbin.fig.suptitle('Hexbin Joint Plot of Total Bill and Tip',
                    fontsize=10, y=1.03)
```

#### 이차원 밀집도 그래프 : kdeplot()

```python
ax = plt.subplots()
ax = sns.kdeplot(x=tips['total_bill'],
                y=tips['tip'],
                shade=True)
ax.set_title('Kernel Density Plot of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
```

```python
ax = plt.subplots()
ax = sns.barplot(x='time', y='total_bill', data=tips)
ax.set_title('Bar plot of average total bill for time of day')
ax.set_xlabel('Time of day')
ax.set_ylabel('Average total bill')
```

### 박스 그래프 : boxplot()

```python
ax = plt.subplots() # 격자 생성
ax = sns.boxplot(x='time', y='total_bill', data=tips)
ax.set_title('Boxplot of total bill by time of day')
ax.set_xlabel('Time of day')
ax.set_ylabel('Total Bill')
```

#### 바이올린 그래프 : violinplot()
- 박스 그래프에서 데이터 분산이 모호하게 표현될 때 사용

```python
ax = plt.subplot()
ax = sns.violinplot(x='time', y='total_bill', data=tips)
ax.set_title('Violin plot of total bill by time of day')
ax.set_xlabel('Time of day')
ax.set_ylabel('Total Bill')
```

### 관계 그래프 : pairplot
- 중복된 그래프가 출력되기 때문에 원하는 그래프로 교체 해줘야함

```python
fig = sns.pairplot(tips)
```

```python
pair_grid = sns.PairGrid(tips)
pair_gird = pair_grid.map_upper(sns.regplot)
pair_grid = pair_grid.map_lower(sns.kdeplot, shade=True)
pair_grid = pair_grid.map_diag(sns.histplot, kde=True)
plt.show()
```

## 데이터프레임과 시리즈로 그래프 나타내기

## seaborn 라이브러리로 그래프 스타일 지정 : set_style()
- 스타일을 적용할 부분에서 한번만 실행

```python
sns.set_style('whitegrid')
fig, ax = plt.subplots()
ax = sns.violinplot(x='time', y='total_bill', hue='sex', data=tips, split=True)
```

```python
# 모든 스타일을 하나씩 적용
fig = plt.figure()
seaborn_styles = ['darkgrid', 'whitegrid', 'dark', 'white', 'ticks']

for idx, style in enumerate(seaborn_styles):
    plot_position = idx + 1
    with sns.axes_style(style):
        ax = fig.add_subplot(2, 3, plot_position)
        violin = sns.violinplot(x='time', y='total_bill', data=tips, ax=ax)
        violin.set_title(style)

fig.tight_layout()
```
